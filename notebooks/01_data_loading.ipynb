```python
# PySpark core
from pyspark.sql import SparkSession

# Spark functions
import pyspark.sql.functions as F
```

Apache Spark digunakan karena kemampuannya dalam memproses data dalam skala besar secara terdistribusi. Spark memungkinkan pemrosesan data dilakukan secara paralel sehingga dapat meningkatkan efisiensi komputasi dibandingkan pendekatan tradisional yang hanya menggunakan satu prosesor. Dengan memanfaatkan SparkSession sebagai titik awal pemrosesan data, seluruh operasi manipulasi dan analisis data dapat dijalankan secara terintegrasi dalam satu lingkungan komputasi.

```python
spark = SparkSession.builder \
    .appName("TMDB Data Loading") \
    .getOrCreate()
```

Kode ini berfungsi untuk mengaktifkan SparkSession, yang merupakan pintu masuk utama untuk menjalankan perintah Apache Spark. Nama aplikasi "TMDB Data Loading" diberikan agar proses pemrosesan data mudah diidentifikasi dalam sistem monitoring. Dengan menggunakan .getOrCreate(), sistem secara cerdas akan membuat sesi baru atau menggunakan sesi yang sudah berjalan jika ada. Dalam konteks Big Data, langkah ini sangat penting karena inilah yang memungkinkan data film yang besar diproses secara paralel di dalam memori, sehingga jauh lebih cepat dibandingkan pengolahan data tradisional.

```python
file_path = "../data/raw/tmdb_movies_raw.parquet"

movies_df = spark.read.parquet(file_path)
```

Karena dataset sudah dalam format Parquet, Spark dapat membaca data dengan sangat cepat.

```python
movies_df.printSchema()
```

Pemeriksaan struktur skema dilakukan untuk memahami tipe data dari setiap atribut yang terdapat dalam dataset.

```python
movies_df.show(5)
```

Langkah ini bertujuan untuk memverifikasi bahwa data telah berhasil dimuat ke dalam lingkungan Apache Spark serta memberikan gambaran awal mengenai struktur dan isi dataset.

```python
movies_df.count()
```

Dataset yang digunakan berisi 9752 film yang dikumpulkan dari platform TMDB. Jumlah film tidak 10.000 karena sudah dilakukan deduplikasi sebelum mengunduh raw dataset.
